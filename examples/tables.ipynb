```python
import numpy as np
from mcdalp.core.credibility import CredibilityMatrix
from mcdalp.core.score import Score
from mcdalp.outranking.crisp import CrispOutranking
```

```python
simple2 = np.array([[1,1,0],[0,1,0],[0,0,0]], dtype=np.int32)
l_simple2 = ["A", "B", "C"]
c_simple2 = CredibilityMatrix(simple2)
s_simple2 = Score()

r_simple2 = CrispOutranking(c_simple2, s_simple2, l_simple2)
r_simple2.solve("complete")

best_r_simple2 = r_simple2.get_rankings()[0]
t_simple2 = best_r_simple2.create_table()
t_simple2.draw()
```
